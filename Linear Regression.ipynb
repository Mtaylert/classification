#### Check out the distribution of your target variable. In this case, our target variable is 'Price'

#### Next step is to split our data

```python
df.columns
```

```python
X = df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']]
y = df['Price']
```

```python
from sklearn.cross_validation import train_test_split
```

#### Pressing Shift + Tab will open up a description window for all python packages/libraries within Jupyter notebook

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101) 
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

#### Evaluate the model

```python
lm.intercept_
```

```python
lm.coef_
```

```python
X_train.columns
```

```python
cdf = pd.DataFrame(lm.coef_,X_train.columns,columns=['Coeff'])
```

```python
cdf
```

### Interpreting the coefficients:
    A 1 unit increase in Avg. Area Income, while holding all else constant, will cause price to increase by $21.528276


### PREDICTIONS

```python
predictions = lm.predict(X_test)
```

```python
predictions
```

```python
y_test.head(5)
```

```python
plt.scatter(y_test,predictions)
```

```python
plt.figure(figsize=(10,10))
sns.distplot((y_test-predictions))
```

### Above we are plotting the distribution of the residuals. A normally distributed residuals plot is a good sign that you picked the right model for the dataset. If it is not normally distributed, you may want to chose a different modeling approach

### Regression Evaluation Metrics
#### Mean Absolute Error (MAE) is the mean of the absolute value of the errors
#### Mean Squared Error (MSE) is the mean of the squared errors
#### Root Mean Squared Error (RMSE) is the square root of the mean of the squared errors


#### MAE is the easiest to understand because it's the average error.
#### MSE is more populare than MAE, because the MSE penalizes larger errors, which tends to be more useful in practice. With this method we are squaring our error
#### RMSE is most populate because it is interpretible in 'y' units

## All of these are known as LOSS FUNCTIONS, because we want to minimize them

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test,predictions))
print('MSE:', metrics.mean_squared_error(y_test,predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test,predictions)))
```

### The two most highly correlated features with our dependent variable look to be LSTAT and RM

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),annot=True)
df.corr()
```

```python
sns.lmplot(x='LSTAT',y='Target',data=df)
```

```python
sns.lmplot(x='RM',y='Target',data=df)
plt.xlabel('Avg. Rooms Per Dwelling')
plt.ylabel('Housing Price')
```

```python
df.columns
```

```python
y=df['Target']
X=df[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX','PTRATIO', 'B', 'LSTAT']]
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101) 
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
lm.intercept_
```

```python
cdf = pd.DataFrame(lm.coef_,X_train.columns,columns=['Coeff'])
```

```python
cdf
```

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
plt.xlabel('Y_test (true values)')
plt.ylabel('Predictions')
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test,predictions))
print('MSE:', metrics.mean_squared_error(y_test,predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test,predictions)))
print('R_Squared:', metrics.explained_variance_score(y_test,predictions))
```

```python
plt.figure(figsize=(10,10))
sns.distplot((y_test-predictions),bins=50)
```
